### Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = list(range(1, 10))
y = [math.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [2 ** n for n in x]
b = [1 / (n + 1) for n in x]


#first subplot
plt.subplot(221)
plt.plot(x, y)
plt.title('x ~ y')

#second subplot
plt.subplot(222)
plt.plot(x, z)
plt.title('x ~ z')

#second subplot
plt.subplot(223)
plt.plot(x, a)
plt.title('x ~ a')

#second subplot
plt.subplot(224)
plt.plot(x, b)
plt.title('x ~ b')

plt.tight_layout()
plt.suptitle('All Four Combined')
plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(1, 4))
y = [math.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [2 ** n for n in x]
b = [1 / (n + 1) for n in x]

plt.figure(figsize=(6, 6)) # (width, height)

plt.plot(x, y, c='red', label='$\sqrt{x}$')
plt.plot(x, z, c='blue', label='labelz')
plt.plot(x, a, c='green', label='labela')
plt.plot(x, b, c='brown', label='labelb')

plt.legend(loc=10)
plt.title('All Four Combined to One')
plt.xlabel('$x$')
plt.ylabel('$y$', rotation=0)

plt.show()
```
